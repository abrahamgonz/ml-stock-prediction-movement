This module was created to extract data from yahoo finance. It is possible to extract only 7 days of intraday data and cannot be older than 30 days.

The Timestamp will be modified to be the same used by BBG.

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import datetime
from datetime import datetime
```

```python
path = 'C:/Users/abrah/Desktop/DISSERTATION/Data/'
```

**The next two functions will be used for the model**

```python
#This option is useful to get information of the last "x" days

def get_data(symbol):
    data = yf.download(tickers=symbol, period="7d", interval="1m")
    data.reset_index(inplace=True)
    data['Datetime'] = pd.to_datetime(data['Datetime']).dt.strftime('%d/%m/%Y %H:%M')
    data.rename(columns = {'Datetime':'Date'}, inplace = True)
    data.to_csv(path + symbol+'_7d.csv', index = False)
    return (data)

#A different option to get data in specific range
#Period can't be longer than 7 days and cannot be older than 30 days
#date must be formatted "2022-06-14"

def get_data_dates(symbol,start_date,end_date):
    data = yf.download(tickers=symbol, start=start_date, end=end_date, interval="1m")
    data.reset_index(inplace=True)
    data['Datetime'] = pd.to_datetime(data['Datetime']).dt.strftime('%d/%m/%Y %H:%M')
    data.rename(columns = {'Datetime':'Date'}, inplace = True)
    data.to_csv(path + symbol+'_'+str(start_date)+'_'+str(end_date)+'.csv')
    return (data) 

def get_data_per_day(symbol,start_date,end_date):
    data = yf.download(tickers=symbol, start=start_date, end=end_date, interval="1m")
    data.reset_index(inplace=True)
    data['Datetime'] = pd.to_datetime(data['Datetime']).dt.strftime('%d/%m/%Y %H:%M')
    data.rename(columns = {'Datetime':'Date'}, inplace = True)
    return (data) 
```

**These are other options to get data**

```python
def get_last_month_data(ticker):    
    dates = pd.date_range(end = datetime.today().strftime('%Y-%m-%d'), periods = 30).to_pydatetime().tolist()
    full_month = pd.DataFrame()
    for i in range(len(dates)-1):
        print(dates[i])
        try:
            new_day = get_data_per_day(ticker,dates[i],dates[i+1]).dropna()
            print(new_day)
            full_month = pd.concat([full_month, new_day])
        except:
            pass
    full_month.to_csv(path + symbol+'_last_month.csv')
    return(full_month)
```
